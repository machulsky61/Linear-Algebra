# Factorización LU

Row echelon pero que no usa intercambio de filas ni multiplicar filas por un escalar no nulo.

```python
def row_echelon_modif(A):
  fils=np.shape(A)[0]
  cols=np.shape(A)[1]
  
  for j in range(cols):
    
    if(A[j,j]!=0):
      for i in range(j+1,fils):
        A[i,:]=A[i,:]+(-A[i,j]/A[j,j])*A[j,:]    
    
    elif(A[j,j]==0 and np.any(A[(j+1):,j]!=0)):
      A=None
      print('Queda un pivote nulo')
      break
  
  return A
```

# Normas y número de condición

Códigos y ejemplos para $v=(1,2,-3)$ y $M=\begin{pmatrix} 3 & 3 \\ -4 & 1 \end{pmatrix}$

```python
v=np.array([1,2,-3])
M=np.array([[3,3],[-4,1]])

nv=np.linalg.norm(v,1)
nm=np.linalg.norm(M,2)
c=np.linalg.cond(M,2)

#Ejercicio: Revisar qué pasa con 'fro', 'np.infty' y p mayor que 1 y distinto de 2#

print(nv)
print(nm)
print(c)
```
